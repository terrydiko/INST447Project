```python
import pandas as pd 
import folium
```

```python

camera = pd.read_csv("https://raw.githubusercontent.com/terrydiko/INST447Project/maCameras.csv", encoding='utf-8')
camera
```

```python
crash = pd.read_csv("https://raw.githubusercontent.com/terrydiko/INST447Project/main/Crashes.csv", encoding= 'utf-8',dtype='string')
crash
```

```python
merge_crash = crash.merge(camera, on='MARID', how='inner')
merge_crash
```

```python
merge_crash[['X', 'Y']]
```

```python
m = folium.Map(location=[38.86948278, -76.96897301], default_zoom_start=200)
folium.Marker(location=[38.86948278, -76.96897301], popup='You found a crash').add_to(m)

m
```
